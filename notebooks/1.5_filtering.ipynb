```python
import numpy as np
import pandas as pd
```

```python
students_performans = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
students_performans = students_performans.rename(columns = {
    'parental level of education': 'parental_level_of_education',
    'test preparation course': 'test_preparation_course',
    'math score': 'math_score',
    'reading score': 'reading_score',
    'writing score': 'writing_score',
})
```

```python
students_performans.head()
```

```python
len(students_performans[students_performans.lunch == 'free/reduced'])/len(students_performans)
```

```python
students_performans_reduced = students_performans[students_performans.lunch == 'free/reduced']
students_performans_standart = students_performans[students_performans.lunch == 'standard']
```

```python
print(students_performans_reduced['math_score'].mean())
print(students_performans_reduced['reading_score'].mean())
print(students_performans_reduced['writing_score'].mean())

print("\n")

print(students_performans_reduced['math_score'].var())
print(students_performans_reduced['reading_score'].var())
print(students_performans_reduced['writing_score'].var())
```

```python
print(students_performans_standart['math_score'].mean())
print(students_performans_standart['reading_score'].mean())
print(students_performans_standart['writing_score'].mean())

print("\n")

print(students_performans_standart['math_score'].var())
print(students_performans_standart['reading_score'].var())
print(students_performans_standart['writing_score'].var())
```
