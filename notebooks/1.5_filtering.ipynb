```python
import numpy as np
import pandas as pd
```

```python
students_performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
students_performance = students_performance.rename(columns = {
    'parental level of education': 'parental_level_of_education',
    'test preparation course': 'test_preparation_course',
    'math score': 'math_score',
    'reading score': 'reading_score',
    'writing score': 'writing_score',
})
```

```python
students_performance.head()
```

```python
len(students_performance[students_performance.lunch == 'free/reduced'])/len(students_performance)
```

```python
students_performance_reduced = students_performance[students_performance.lunch == 'free/reduced']
students_performance_standart = students_performance[students_performance.lunch == 'standard']
```

```python
print(students_performance_reduced['math_score'].mean())
print(students_performance_reduced['reading_score'].mean())
print(students_performance_reduced['writing_score'].mean())

print("\n")

print(students_performance_standart['math_score'].var())
print(students_performance_standart['reading_score'].var())
print(students_performance_standart['writing_score'].var())
```

```python
print(students_performance_standart['math_score'].mean())
print(students_performance_standart['reading_score'].mean())
print(students_performance_standart['writing_score'].mean())

print("\n")

print(students_performance_standart['math_score'].var())
print(students_performance_standart['reading_score'].var())
print(students_performance_standart['writing_score'].var())
```
