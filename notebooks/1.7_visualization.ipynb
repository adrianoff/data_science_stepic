```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sb

import numpy as np
import pandas as pd
```

```python
students_performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
students_performance = students_performance.rename(columns = {
    'parental level of education': 'parental_level_of_education',
    'test preparation course': 'test_preparation_course',
    'math score': 'math_score',
    'reading score': 'reading_score',
    'writing score': 'writing_score',
})
```

```python
students_performance.head()
```

```python
students_performance.math_score.hist()
```

```python
students_performance.plot.scatter(x='math_score', y='reading_score')
```

```python
sb.lmplot(x='math_score', y='reading_score', data=students_performance)
```

```python
ax = sb.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance)
ax.set_xlabels('Math Score')
ax.set_ylabels('Reading Score')
```
